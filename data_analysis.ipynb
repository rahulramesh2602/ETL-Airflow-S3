```python
#Importing necessary packages
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Twitter_data.csv")
```

```python
df.head()
```

```python
df = df.drop(["country","id","language","latitude","longitude"], axis=1)
df.head()
```

```python
df = df.rename(columns= {"author": "user", "content": "text", "number_of_likes": "favourite_count", "number_of_shares": "retweet_count","date_time": "created_at"})
df.head()
```

```python
df.to_csv("Twitter_data.csv", index= False)
```
